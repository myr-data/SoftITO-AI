## Data Wrangling

The null values are done away with *(since there wasn't much already of them, I simply removed them, but this is normally NOT recommended unless the entire column is missing more than 40% of its data)*, and then, we can focus on what else may be wrong.

**Normally**
- Replacement with Mean/Median/Mode
- Random Sample Imputation
- Capturing NAN with new feature
- End of Distribution Imputation
- Arbitrary Imputation
- Frequent Category Imputation (only used with Categorical feature)
- Make ‘NAN’ a new Category(only used with Categorical feature)

**Depends on**
- MCAR: means Missing completely at random. The probability of a missing data value is independent of any observation in the data set means it has no relation with the data, it just miss at random places.
- MNAR: means Missing Not at Random. For example, when most of the missing people from work are sickest people, people with the lowest education are missing on education, this kind of missing is referred as Missing Not at Random (MNAR).
- MAR: means Missing at Random

### Categorical Columns
Linear Regression works with numerical values. So we transform categorical (text based) values into numerical values, without making them lose their meaning or relationships.

```python
print("Show categorical columns and the contents")
print("-"*20)
for col in df.select_dtypes(include=['object']):
    print(f"Column: {col}")
    print(df[col].value_counts())
    print("\n")
```

### Mapper
Mappers describe which text turns into which value.

```python
lowmedhigh_mapper = {
    "Low": 0,
    "Medium": 1,
    "High": 2
}

schooltype_mapper = {
    "Public": 0,
    "Private": 1
}

yesno_mapper = {
    "Yes": 1,
    "No": 0
}

posneutneg_mapper = {
    "Positive": 2,
    "Neutral": 1,
    "Negative": 0
}

education_mapper = {
    "High School": 0,
    "College": 1,
    "Postgraduate": 2
}

distance_mapper = {
    "Far": 0,
    "Moderate": 1,
    "Near": 2
}

gender_mapper = {
    "Female": 0,
    "Male": 1
}
```

```python
for col in [
    "School_Type",
    "Peer_Influence",
    "Parental_Education_Level",
    "Distance_from_Home",
    "Gender"
]:
    print(col)
    print(df[col].unique())
    print()
```

```python
df.select_dtypes(include=["object"]).columns
```

```python
lowhighmed = [
    "Parental_Involvement",
    "Access_to_Resources",
    "Motivation_Level",
    "Family_Income",
    "Teacher_Quality"
]

yesno = [
    "Learning_Disabilities",
    "Internet_Access",
    "Extracurricular_Activities"
]
```

```python
for col in lowhighmed:
    df[col] = df[col].map(lowmedhigh_mapper)

for col in yesno:
    df[col] = df[col].map(yesno_mapper)
```

```python
df["Gender"] = df["Gender"].map(gender_mapper)
df["School_Type"] = df["School_Type"].map(schooltype_mapper)
df["Peer_Influence"] = df["Peer_Influence"].map(posneutneg_mapper)
df["Parental_Education_Level"] = df["Parental_Education_Level"].map(education_mapper)
df["Distance_from_Home"] = df["Distance_from_Home"].map(distance_mapper)
```

```python
df.select_dtypes(include=["object"]).columns
```

```python
print("After mapping is done")
print("-"*20)
df.info()
```

```python
print("Columns to use in guessing (X) and column that the model will try to guess (y)")
print("-"*20)
X = df.drop("Exam_Score", axis=1)
y = df["Exam_Score"]

# Boyutları kontrol etme
print(f"X shape:{X.shape}")
print(f"y shape: {y.shape}")
```

```python
print("Dataset saved as student_eda.csv in savepath")
df.to_csv(savepath + r"/student_eda.csv", index=False)
```

```python
print("Split X and y into training and testing subsets")
print("-"*20)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
print("Scale the data based on standard deviation")
print("-"*20)

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
print("Fit training data into linear regression model")
print("-"*20)

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
print("Predictions done")
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
print("Evalution metrics")
print("-"*20)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)

print(f"MAE : {mae:.3f}")
print(f"MSE : {mse:.3f}")
print(f"RMSE: {rmse:.3f}")
print(f"R²  : {r2:.3f}")
```

```python
coefficients = (
    pd.DataFrame({
        "Feature": X.columns,
        "Coefficient": model.coef_
    })
    .sort_values("Coefficient", key=abs, ascending=False)
)
print("How each feature affects the target")
print("-"*20)

print(coefficients)
```
