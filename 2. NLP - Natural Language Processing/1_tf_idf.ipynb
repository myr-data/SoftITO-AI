```python
text1 = "Double Normalization Term Frequency değeri ile IDF değerinin çarpımı bize bu terimin doküman içerisindeki değerini vermiş olur."
text2 = "Dokümanın ağırlıklı olarak hangi konuyla ilgili olduğunu, arama işlemlerinde aradığımız terimin dokümanda ne kadar ağırlıklı olarak geçtiğini anlayabiliriz. Lucene dokümanları indeksleyen ve bu dokümanlar üzerinde yüksek performanslı kelime araması yapabildiğiniz bir java kütüphanesidir."
text3 = "Özetle terim ne kadar az dökumanda tekrar ediyor ise IDF değeri o kadar büyük çıkar."
docs = [text1, text2, text3]
```

```python
class TextProcess():
  def __init__(self, docs):
    self.docs=docs
  def tokenize_docs(self):
    newdoc=[]
    for d in self.docs:
      newdoc.append(d.lower().replace(".","").replace(",","").split())
    for i, d in enumerate(newdoc):
      print(f"Document {i+1}: {d}")
    return newdoc
```

```python
tp = TextProcess(docs)
newdoc = tp.tokenize_docs()
```

```python
class TF_IDF():
  def __init__(self, tokenized_docs):
    self.tokenized_docs = tokenized_docs

  def tf(self, term, doc_tokens):
    return doc_tokens.count(term)

  def idf(self, term):
    doc_frequency = 0
    for doc_tokens in self.tokenized_docs:
      if term in doc_tokens:
        doc_frequency += 1
    if doc_frequency == 0:
      return 0
    return math.log(len(self.tokenized_docs) / doc_frequency)

  def tf_idf(self):
    for i, doc_tokens in enumerate(self.tokenized_docs):
      print(f"Document {i+1}:")
      print()
      print(f"Tokenized: {doc_tokens}")
      unique_terms = sorted(set(doc_tokens))
      for term in unique_terms:
        tf_val = self.tf(term, doc_tokens)
        idf_val = self.idf(term)
        print(f"{term}: TF={round(tf_val,3)}, IDF={round(idf_val,3)}, TF-IDF={round(tf_val * idf_val,3)}")
      print()
```

```python
tfidf = TF_IDF(newdoc)
tfidf.tf_idf()
```

```python
# En önemli 3 kelimeyi bulma
all_terms = []
for doc_tokens in newdoc:
    all_terms.extend(doc_tokens)

term_scores = {}
for term in set(all_terms):
    term_scores[term] = sum(
        (doc_tokens.count(term) * tfidf.idf(term)) for doc_tokens in newdoc
    )

important_terms = sorted(term_scores.items(), key=lambda x: x[1], reverse=True)[:3]

print("En önemli 3 kelime:")
for term, score in important_terms:
    print(f"{term}: {score:.3f}")
```
